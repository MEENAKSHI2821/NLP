```python
#Create dataframe for all the speeches

filelist= os.listdir()
data = []
for f in filelist:
    with open(f,'r') as myfile:
        string = myfile.read()
        data.append(string)
```

```python
########### concatenate speeches based on FM
PM_txt= data[0]+ data[1]+data[2]

PC_txt= data[3]+data[4]

AJ_txt= data[5]+data[6]+data[7]+data[8]+data[9]

PG_txt= data[10]

NS_txt= data[11]
```

```python
########### Preliminary cleaning

PM_txt =  PM_txt.replace('\n\n','\n')
PM_txt =  PM_txt.replace('\n','')
PM_txt =  PM_txt.strip()
PM_txt =  re.sub(' +', ' ',PM_txt)
PM_txt =  re.sub('\.',' ',PM_txt)
PM_txt =  PM_txt.lower()
PM_txt =  re.sub("[0-9]"," ",PM_txt)
PM_txt =  re.sub("\s+\d+\W+", "", PM_txt)
PM_txt =  re.sub( r'\([a-z]+\)' ," ",PM_txt)
PM_txt = ' '.join([w for w in PM_txt.split() if len(w)>3])

import string
for c in string.punctuation:
    PM_txt= PM_txt.replace(c,"")

PC_txt =  PC_txt.replace('\n\n','\n')
PC_txt =  PC_txt.replace('\n','')
PC_txt =  PC_txt.strip()
PC_txt =  re.sub(' +', ' ',PC_txt)
PC_txt =  re.sub('\.',' ',PC_txt)
PC_txt =  PC_txt.lower()
PC_txt =  re.sub("[0-9]"," ",PC_txt)
PC_txt =  re.sub("\s+\d+\W+", "", PC_txt)
PC_txt =  re.sub( r'\([a-z]+\)' ," ",PC_txt)
PC_txt = ' '.join([w for w in PC_txt.split() if len(w)>3])

import string
for c in string.punctuation:
    PC_txt= PC_txt.replace(c,"")


AJ_txt =  AJ_txt.replace('\n\n','\n')
AJ_txt =  AJ_txt.replace('\n','')
AJ_txt =  AJ_txt.strip()
AJ_txt =  re.sub(' +', ' ',AJ_txt)
AJ_txt =  re.sub('\.',' ',AJ_txt)
AJ_txt =  AJ_txt.lower()
AJ_txt =  re.sub("[0-9]"," ",AJ_txt)
AJ_txt =  re.sub("\s+\d+\W+", "", AJ_txt)
AJ_txt =  re.sub( r'\([a-z]+\)' ," ",AJ_txt)
AJ_txt = ' '.join([w for w in AJ_txt.split() if len(w)>3])

import string
for c in string.punctuation:
    AJ_txt= AJ_txt.replace(c,"")

PG_txt =  PG_txt.replace('\n\n','\n')
PG_txt =  PG_txt.replace('\n','')
PG_txt =  PG_txt.strip()
PG_txt =  re.sub(' +', ' ',PG_txt)
PG_txt =  re.sub('\.',' ',PG_txt)
PG_txt =  PG_txt.lower()
PG_txt =  re.sub("[0-9]"," ",PG_txt)
PG_txt =  re.sub("\s+\d+\W+", "", PG_txt)
PG_txt =  re.sub( r'\([a-z]+\)' ," ",PG_txt)
PG_txt = ' '.join([w for w in PG_txt.split() if len(w)>3])

import string
for c in string.punctuation:
    PG_txt= PG_txt.replace(c,"")

NS_txt =  NS_txt.replace('\n\n','\n')
NS_txt =  NS_txt.replace('\n','')
NS_txt =  NS_txt.strip()
NS_txt =  re.sub(' +', ' ',NS_txt)
NS_txt =  re.sub('\.',' ',NS_txt)
NS_txt =  NS_txt.lower()
NS_txt =  re.sub("[0-9]"," ",NS_txt)
NS_txt =  re.sub("\s+\d+\W+", "", NS_txt)
NS_txt =  re.sub( r'\([a-z]+\)' ," ",NS_txt)
NS_txt = ' '.join([w for w in NS_txt.split() if len(w)>3])

import string
for c in string.punctuation:
    NS_txt= NS_txt.replace(c,"")
```

```python
#######Tokenization for Fdist and graphs

from nltk.tokenize import word_tokenize
PM_txt_Tok = word_tokenize(PM_txt.lower())

from nltk.tokenize import word_tokenize
PC_txt_Tok = word_tokenize(PC_txt.lower())

from nltk.tokenize import word_tokenize
AJ_txt_Tok = word_tokenize(AJ_txt.lower())

from nltk.tokenize import word_tokenize
PG_txt_Tok = word_tokenize(PG_txt.lower())

from nltk.tokenize import word_tokenize
NS_txt_Tok = word_tokenize(NS_txt.lower())

#####################

from nltk.probability import FreqDist
PM_txt_Tok_dist = FreqDist(PM_txt_Tok)

from nltk.probability import FreqDist
PC_txt_Tok_dist = FreqDist(PC_txt_Tok)

from nltk.probability import FreqDist
AJ_txt_Tok_dist = FreqDist(AJ_txt_Tok)

from nltk.probability import FreqDist
PG_txt_Tok_dist = FreqDist(PG_txt_Tok)

from nltk.probability import FreqDist
NS_txt_Tok_dist = FreqDist(NS_txt_Tok)

```

```python
#########Creating DF for BIGRAMS
#########Creating DF for BIGRAMS
PM_txt_Tok_Stop_str_DF=pd.DataFrame([PM_txt_Tok_Stop_str])
PM_txt_Tok_Stop_str_DF.columns= ['Speech']

PC_txt_Tok_Stop_str_DF=pd.DataFrame([PC_txt_Tok_Stop_str])
PC_txt_Tok_Stop_str_DF.columns= ['Speech']

AJ_txt_Tok_Stop_str_DF=pd.DataFrame([AJ_txt_Tok_Stop_str])
AJ_txt_Tok_Stop_str_DF.columns= ['Speech']

PG_txt_Tok_Stop_str_DF=pd.DataFrame([PG_txt_Tok_Stop_str])
PG_txt_Tok_Stop_str_DF.columns= ['Speech']

NS_txt_Tok_Stop_str_DF=pd.DataFrame([NS_txt_Tok_Stop_str])
NS_txt_Tok_Stop_str_DF.columns= ['Speech']
```

```python
######## CREATE BIGRAMS and Trigrams

from sklearn.feature_extraction.text import CountVectorizer
bigram_count_vectorizer = CountVectorizer(ngram_range=(2,2),max_features=150)

from sklearn.feature_extraction.text import CountVectorizer
trigram_count_vectorizer = CountVectorizer(ngram_range=(3,3),max_features=150)

#################
X_bigram_PM = bigram_count_vectorizer.fit_transform(PM_txt_Tok_Stop_str_DF['Speech'])
DTM_bigram_PM = pd.DataFrame(X_bigram_PM.toarray(), columns=bigram_count_vectorizer.get_feature_names())

X_trigram_PM = trigram_count_vectorizer.fit_transform(PM_txt_Tok_Stop_str_DF['Speech'])
DTM_trigram_PM = pd.DataFrame(X_trigram_PM.toarray(), columns=trigram_count_vectorizer.get_feature_names())

######

X_bigram_PC = bigram_count_vectorizer.fit_transform(PC_txt_Tok_Stop_str_DF['Speech'])
DTM_bigram_PC = pd.DataFrame(X_bigram_PC.toarray(), columns=bigram_count_vectorizer.get_feature_names())

X_trigram_PC = trigram_count_vectorizer.fit_transform(PC_txt_Tok_Stop_str_DF['Speech'])
DTM_trigram_PC = pd.DataFrame(X_trigram_PC.toarray(), columns=trigram_count_vectorizer.get_feature_names())

######
X_bigram_AJ = bigram_count_vectorizer.fit_transform(AJ_txt_Tok_Stop_str_DF['Speech'])
DTM_bigram_AJ = pd.DataFrame(X_bigram_AJ.toarray(), columns=bigram_count_vectorizer.get_feature_names())

X_trigram_AJ = trigram_count_vectorizer.fit_transform(AJ_txt_Tok_Stop_str_DF['Speech'])
DTM_trigram_AJ = pd.DataFrame(X_trigram_AJ.toarray(), columns=trigram_count_vectorizer.get_feature_names())

######
X_bigram_PG = bigram_count_vectorizer.fit_transform(PG_txt_Tok_Stop_str_DF['Speech'])
DTM_bigram_PG = pd.DataFrame(X_bigram_PG.toarray(), columns=bigram_count_vectorizer.get_feature_names())

X_trigram_PG = trigram_count_vectorizer.fit_transform(PG_txt_Tok_Stop_str_DF['Speech'])
DTM_trigram_PG = pd.DataFrame(X_trigram_PG.toarray(), columns=trigram_count_vectorizer.get_feature_names())

######
X_bigram_NS = bigram_count_vectorizer.fit_transform(NS_txt_Tok_Stop_str_DF['Speech'])
DTM_bigram_NS = pd.DataFrame(X_bigram_NS.toarray(), columns=bigram_count_vectorizer.get_feature_names())

X_trigram_NS = trigram_count_vectorizer.fit_transform(NS_txt_Tok_Stop_str_DF['Speech'])
DTM_trigram_NS = pd.DataFrame(X_trigram_NS.toarray(), columns=trigram_count_vectorizer.get_feature_names())

```
